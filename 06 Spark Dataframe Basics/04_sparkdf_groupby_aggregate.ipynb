```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('aggs').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/sales_info.csv',inferSchema=True,header=True)
```

```
df.show()
```

```
df.printSchema()
```

```
df.groupBy('Company').mean().show()
#can do sum(), max(), min(),count() and many more
```

```
#If you want everything aggregated:
df.agg({'Sales':'sum'}).show()
#similarly max,min,avg,count
```

```
#Groupby + Agg
group_data = df.groupBy('Company')
group_data.agg({'Sales':'mean'}).show()
```

```
#import functions from Spark
from pyspark.sql.functions import countDistinct,avg,stddev
```

```
df.select(avg('Sales')).show()
```

```
df.select(avg('Sales').alias('Average Sales')).show()
```

```
df.select(stddev('Sales')).show()
```

```
from pyspark.sql.functions import format_number
```

```
sales_std = df.select(stddev('Sales').alias('Sales Standard Deviation'))
sales_std.select(format_number('Sales Standard Deviation',2).alias('Sales Standard Deviation')).show()
```

```
#Sorting
# df.orderBy('Sales').show()
df.orderBy(df['Sales'].desc()).show()
```
