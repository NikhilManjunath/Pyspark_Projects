```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('dates').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/appl_stock.csv',inferSchema=True,header=True)
```

```
df.show()
```

```
df.select(['Date','Open']).show()
```

```
from pyspark.sql.functions import dayofmonth,hour,dayofyear,month,year,weekofyear,format_number,date_format
```

```
df.select(dayofmonth(df['Date'])).show()
```

```
df.select(month(df['Date'])).show()
```

```
new_df = df.withColumn('Year',year(df['Date']))
new_df.show()
```

```
#If we want to check average closing price by year
result = new_df.groupBy('Year').mean().select(['Year','avg(Close)'])
result.show()
```

```
result.withColumnRenamed('avg(Close)','Average Closing Price').orderBy('Year').select(['Year',format_number('Average Closing Price',2).alias('Closing Price')]).show()
```
