```python
#Changing the schema
from pyspark.sql.types import StructField,StringType,IntegerType,StructType
```

```python
#Creating a schema
data_schema = [StructField('age',IntegerType(),True),
              StructField('name',StringType(),True)]
```

```python
#Creating a variable of StructType with fields set to the created schema
final_struc = StructType(fields=data_schema)
```

```python
#Reading the JSON file by specifying our schema
df = spark.read.json('/FileStore/tables/people.json',schema=final_struc)
```

```python
df.printSchema()
```

```python
df.show()
```

```python
#Grabbing data
#Selecting only the required column
#This just returns a column
df['age']
```

```python
#Selecting only the required column
#This returns a df of the column
df.select('age').show()
```

```python
#Selecting Multiple columns
df.select(['age','name']).show()
```

```python
#Getting row entries
#This returns a list of row objects
df.head(2)
```

```python
#Creating new columns
df.withColumn('double_age',df['age']*2).show()
#will have to do inplace to save to original df
```

```python
#Renaming a column
df.withColumnRenamed(existing='age',new='new_age').show()
```

```python
#Using SQL to get data
df.createOrReplaceTempView('people')
```
