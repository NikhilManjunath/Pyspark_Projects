```
from pyspark.sql import SparkSession
sparrk = SparkSession.builder.appName('log_reg').getOrCreate()
```

```
from pyspark.ml.classification import LogisticRegression
```

```
my_data = spark.read.format('libsvm').load('/FileStore/tables/sample_libsvm_data.txt')
my_data.show()

lr_train,lr_test = my_data.randomSplit([0.7,0.3])
```

```
log_reg = LogisticRegression()
log_reg_model = log_reg.fit(lr_train)
```

```
lr_test_preds = log_reg_model.evaluate(lr_test)
lr_test_preds.predictions.show()
```

```
lr_test_preds.accuracy
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator,MulticlassClassificationEvaluator
```

```
evaluator = BinaryClassificationEvaluator()
roc_score = evaluator.evaluate(lr_test_preds.predictions)
print('ROC Score: {0}'.format(roc_score))
```
