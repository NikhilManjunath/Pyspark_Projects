## Bonus Question!
#### There are too many decimal places for mean and stddev in the describe() dataframe. Format the numbers to just show up to two decimal places. Pay careful attention to the datatypes that .describe() returns, we didn't cover how to do this exact formatting, but we covered something very similar. [Check this link for a hint](http://spark.apache.org/docs/latest/api/python/pyspark.sql.html#pyspark.sql.Column.cast)

If you get stuck on this, don't worry, just view the solutions.

```python
summary = df.describe()
summary.printSchema()
```

```python
summary_df = summary.select(['summary',
                             format_number(summary['Open'].cast('float'),2).alias('Open'),
                             format_number(summary['High'].cast('float'),2).alias('High'),
                             format_number(summary['Low'].cast('float'),2).alias('Low'),
                             format_number(summary['Close'].cast('float'),2).alias('Close'),
                             format_number(summary['Volume'].cast('int'),2).alias('Volume')])
summary_df.show()
```

#### Create a new dataframe with a column called HV Ratio that is the ratio of the High Price versus volume of stock traded for a day.

```python
df.withColumn('HV Ratio',df['High']/df['Volume']).select(['HV Ratio']).show()
```

#### What day had the Peak High in Price?

```python
peak_high = df.agg({'High':'max'}).collect()
peak_high_date = df.filter(df['High'] == peak_high[0][0]).select(['Date'])
peak_high_date.show()
```

#### What is the mean of the Close column?

```python
df.agg({'Close':'mean'}).show()
```

#### What is the max and min of the Volume column?

```python
from pyspark.sql.functions import max,min
df.select(max('Volume'),min('Volume')).show()
```

#### How many days was the Close lower than 60 dollars?

```python
from pyspark.sql.functions import count
df.filter(df['Close'] < 60).count()
```

#### What percentage of the time was the High greater than 80 dollars ?
#### In other words, (Number of Days High>80)/(Total Days in the dataset)

```python
df.filter(df['High'] > 80).count()*100/df.count()
```

#### What is the Pearson correlation between High and Volume?
#### [Hint](http://spark.apache.org/docs/latest/api/python/pyspark.sql.html#pyspark.sql.DataFrameStatFunctions.corr)

```python
from pyspark.sql.functions import corr
```

```python
df.select(corr('High','Volume')).show()
```

#### What is the max High per year?

```python
from pyspark.sql.functions import dayofmonth,hour,dayofyear,month,year,weekofyear,format_number,date_format
```

```python
new_df = df.select(year(df['Date']).alias('Year'),'High')
new_df.groupBy('Year').max().select(['Year','max(High)']).show()
```

#### What is the average Close for each Calendar Month?
#### In other words, across all the years, what is the average Close price for Jan,Feb, Mar, etc... Your result will have a value for each of these months.

```python
new_df = df.select(month(df['Date']).alias('Month'),'Close')
new_df.groupBy('Month').mean().select(['Month','avg(Close)']).orderBy(new_df['Month']).show()
```
